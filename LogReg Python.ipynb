```python
data1=data
data1
```

```python
#we are doing labelencoder for the ID's column as they are in ordinal scale
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Encode single column statusdata.StageID = le.fit_transform(data.StageID)
data1.StageID = le.fit_transform(data1.StageID)
data1.GradeID = le.fit_transform(data1.GradeID)
data1.SectionID = le.fit_transform(data1.SectionID)

```

```python
data1.Class.value_counts()
```

```python
data1.Class.replace({'M':0},inplace=True)
data1.Class.replace({'H':1},inplace=True)
data1.Class.replace({'L':2},inplace=True)
#we are replacing the values for class according to their counts 
```

```python
data1
```

```python
data1.info()
```

```python
dep=data1['Class']
dep
```

```python
data1=data1.drop(['Class'],axis=1)
```

```python
data1=pd.get_dummies(data1)
data1
```

```python
#we are doing one-hot encoder for all the independent variables as they are in nominal scale
```

```python
idv=data1
idv
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=idv
y=dep
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.3)
```

```python
Prevalance_rate_ytrain=ytrain.value_counts(normalize=True).mul(100).round(1)
Prevalance_rate_ytrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=5000)
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtrain)
ypred
```

```python
ypred1=model.predict(xtest)
ypred1
```

```python
ytrainpred=model.predict_proba(xtrain)
ytrainpred
```

```python
ytestpred=model.predict_proba(xtest)
ytestpred
```

```python
np.argmax(ytrainpred>0.7,axis=1)
```

```python
np.argmax(ytestpred>0.6,axis=1)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(ytrain,ypred)  # we have diff only within 0.5
```

```python
accuracy_score(ytest,ypred1)
```

```python
confusion_matrix(ytrain,ypred)
```

```python
confusion_matrix(ytest,ypred1)
```

```python
classification_report(ytrain,ypred)
```
